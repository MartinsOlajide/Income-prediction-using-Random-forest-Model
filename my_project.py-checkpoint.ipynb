```python
frequency = df['Age'].value_counts()
```

```python
percentage = df['Age'].value_counts(normalize=True).mul(100).round(1).astype(str).add('%')
```

```python
result_df1 = pd.DataFrame({
    'Age(years)': frequency.index,
    'Frequency': frequency.values,
    'Percentage': percentage.values
        })
```
